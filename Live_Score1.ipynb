```python
from bs4 import BeautifulSoup
import requests
import urllib
import matplotlib.pyplot as plt 
import lxml
```

```python
url1="http://scores.sify.com"
src=requests.get(url1)
```

```python
soup=BeautifulSoup(src.content,'lxml')
```

```python
count=0
matches=[]
for i in soup.find_all('h2'):
    count+=1
    if count==4:
        break
    matches.append(i.text)
print(f'Matches that are today are:{matches}')
```

```python
url=[]
for i in soup.find_all("div",class_="scoresbox-center"):
    try:
        url.append(i.h2.a.get('href'))
    except:
        break
url
```

```python
for i in soup.find_all('div',class_="scoresbox-center"):
    
    try:
        p=i.text




        plot=[]
        for p in i.strings:
            plot.append(p)
        print(f'{plot[0]}\n{plot[1]} {plot[3]} {plot[5]}\n{plot[6]} score: {plot[7]}\n{p}')

        #finding batsman
        p=[]
        for k in range(3,5):
            p.append(i.find_all('a')[k].text)


        print(f'Batsman are {p}')

    except:
        pass
    print('\n--------------------------------------------')
    
    
```
